# 05. Data Consistency

## Missing values:

```python
# Check for missing values
df_merged.isnull().sum()
```

### There seems to be some missing values in the Region column als well as the previously identified missing value in the Perceptions of Corruption column.

```python
# Checking to see if the missing value is still the same 

df_merged[df_merged['Perceptions of Corruption'].isnull()]
```

### The value in this column is still the previously identified one.

```python
# Checking the mising values in the Region column

print(df_merged[df_merged['Region'].isnull()][['Country', 'Year']])
pd.set_option('display.max_rows', None)
```

### After checking the missing values it is clear that there are multiple reasons why the Region column might be missing for a country:  

### 1. The name of the country has been changed for different years (Taiwan, Northern Cyprus, Trinidad & Tobago, Hongkong, Czechia, State of Palestine, Turkiye).

### 2.  An asterik has been added to a country ("*"), so the country names dont match up.

### 3. The country is only in present in some of the years and not in the years where a region is assigned to the countries (2020 & 2021).

```python
# Changing the country names to a unique format
df_merged['Country'] = df_merged['Country'].replace({
    'Taiwan Province of China': 'Taiwan',
    'Northern Cyprus': 'North Cyprus',
    'Trinidad & Tobago': 'Trinidad and Tobago',
    'Hong Kong S.A.R. of China': 'Hong Kong',
    'Czechia': 'Czech Republic',
    'Palestinian Territories': 'State of Palestine',
    'Turkey': 'Turkiye'
}, regex=True)
```

```python
# Drop the asterik from the 2022 Region column

df_merged['Country'] = df_merged['Country'].str.replace('*', '', regex=False)
```

```python
# Fill in the missing Region columns again
df_merged['Region'] = df_merged.groupby('Country')['Region'].transform(lambda x: x.ffill().bfill())
```

```python
# Checking the mising values in the Region column

print(df_merged[df_merged['Region'].isnull()][['Country', 'Year']])
pd.set_option('display.max_rows', None)
```

```python
# Change the name of Eswatini
df_merged['Country'] = df_merged['Country'].replace('Eswatini, Kingdom of', 'Eswatini')
```

```python
# Assign a Region to the remaining countries
df_merged.loc[df_merged['Country'] == 'Qatar', 'Region'] = 'Middle East and North Africa'
df_merged.loc[df_merged['Country'] == 'Belize', 'Region'] = 'Latin America and Caribbean'
df_merged.loc[df_merged['Country'] == 'Bhutan', 'Region'] = 'South Asia'
df_merged.loc[df_merged['Country'] == 'Somalia', 'Region'] = 'Sub-Saharan Africa'
df_merged.loc[df_merged['Country'] == 'Sudan', 'Region'] = 'Sub-Saharan Africa'
df_merged.loc[df_merged['Country'] == 'Angola', 'Region'] = 'Sub-Saharan Africa'
df_merged.loc[df_merged['Country'] == 'Syria', 'Region'] = 'Middle East and North Africa'
df_merged.loc[df_merged['Country'] == 'Congo', 'Region'] = 'Sub-Saharan Africa'
df_merged.loc[df_merged['Country'] == 'Eswatini', 'Region'] = 'Sub-Saharan Africa'
```

```python
print(df_merged[df_merged['Region'].isnull()][['Country', 'Year']])
```

### There are two different countries named congo in the dataframe, distinguished by their capital in brackets. The entry for 2022 is only included as "congo". By checking the rankings of the other years it is evident that this has to be an entry for country 'Congo (Brazzaville)'.

```python
# Rename the countries for clarity 

df_merged.loc[df_merged['Country'] == 'Congo (Brazzaville)', 'Country'] = 'Republic of the Congo'


df_merged.loc[df_merged['Country'] == 'Congo', 'Country'] = 'Republic of the Congo'


df_merged.loc[df_merged['Country'] == 'Congo (Kinshasa)', 'Country'] = 'Democratic Republic of the Congo'
```
